Затем я начал подбирать модель. По итогам тестов (ооооочень долгих) я выяснил, что лучшее качество получается, если использовать стемминг Портера, после него использовать CountVectorizer без стоп слов, с униграммами и анализатором word, а после этого отсеивать все незначимые признаки через корелляционный фильтр на основе корреляции Мэтьюса с порогом 0.023. Так же оптимальный параметр alpha для модели оказался равен 1.1

Я написал классы, соответствующие этим шагам, чтобы собрать пайплайн. 

```python
#Transformer class based on correlation filter
class CorrTransformer(TransformerMixin):
    def __init__(self, corr_limit=0.01, correlation='pearson'):
        if correlation not in ['pearson', 'matthews']:
            raise ValueError('Correaltion must have "pearson" or "matthews" value')
        self.corr_limit = corr_limit
        self.corr = correlation
        self.coefs = []
        
    def __matthews_corr__(self, arr1, arr2):
        if len(arr1) != len(arr2):
            raise ValueError('lenths of arrays must be same')
        from math import sqrt
        a, b, c, d = 0, 0, 0, 0
        for e, i in zip(arr1, arr2):
            if e == 0 and i == 0:
                a += 1
            elif e == 0 and i == 1:
                b += 1
            elif e != 0 and i == 0:
                c += 1
            elif e != 0 and i == 1:
                d += 1
        if a + b + c + d != len(arr1):
            raise ValueError('Counting Error')
        S = (a + b) / len(arr1)
        P = (a + c) / len(arr1)
        N = len(arr1)
        return ((a / N) - S * P) / sqrt(P * S * (1 - P) * (1 - S))
        
    def fit(self, data, labels, *args):
        corcoefs = []
        if self.corr == 'pearson':
            corcoefs = [np.corrcoef(data[:, i].toarray().ravel(), labels)[0, 1] 
                        for i in range(data.shape[1])]
        elif self.corr == 'matthews':
            corcoefs = [self.__matthews_corr__(data[:, i].toarray().ravel(), labels) 
                        for i in range(data.shape[1])]
        self.coefs = np.array(corcoefs)
        return self
    
    def transform(self, data, *args):
        indexes = [i for i in range(len(self.coefs)) if abs(self.coefs[i]) > self.corr_limit]
        return data[:, indexes]
```

```python
class StemmerTransformer(TransformerMixin):
    def __init__(self, stemmer='snowball'):
        if stemmer not in ['snowball', 'lancaster', 'porter']:
            raise ValueError('stemmer must have showball, lancaster or porter value')
        self.stemmer = stemmer
    
    def fit(self, *args):
        return self
    
    def transform(self, data, *args):
        if self.stemmer == 'snowball':
            from nltk.stem import SnowballStemmer
            stemmer = SnowballStemmer(language='english')
        elif self.stemmer == 'lancaster':
            from nltk.stem.lancaster import LancasterStemmer
            stemmer = LancasterStemmer()
        elif self.stemmer == 'porter':
            from nltk.stem.porter import PorterStemmer
            stemmer = PorterStemmer()
        new_data = []
        for doc in data:
            clear_doc = re.split(r'[\s]',  doc)
            words = [stemmer.stem(word.strip()) for word in clear_doc if word.strip() != '']
            new_data.append(' '.join(words))
        return np.array(new_data)
```

Так же я пробовал использовать лемматизацию, но качетсво от этого упало

```python
class LemmatizerTransformer():
    def __init__(self):
        from nltk.stem import WordNetLemmatizer
        self.lemmatizer = WordNetLemmatizer()
    def fit(self, data, labels):
        return self
    def transform(self, data):
        new_data = []
        for doc in data:
            clear_doc = re.split(r'[\s]',  str(doc))
            words = [self.lemmatizer.lemmatize(word.strip()) for word in clear_doc if word.strip() != '']
            new_data.append(' '.join(words))
        return np.array(new_data)
```

```python
pl = Pipeline([('stem', StemmerTransformer(stemmer='porter')),
               ('vect', CountVectorizer(ngram_range=(1,1), analyzer='word')), 
               ('transformer', CorrTransformer(corr_limit=0.023, correlation='matthews')),
               ('model', MultinomialNB(alpha=1.1))])
result = cross_val_score(pl, data, labels, cv=3).mean()
print(result)
```
